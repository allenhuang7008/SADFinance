```python
import yfinance as yf
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
data = "/Users/hp/SAD_Finance/nasdaq_screener_1686545341556.csv"
stock=pd.read_csv(data)
len(stock)
```

```python
stock1 = stock[0:1000]
stock1
stock_symbol1 = stock1["Symbol"]
symbol1 = stock_symbol1.tolist()

test_symbol = ['A','AA','AAC','AACG','AACI','AACIU','AACIW','AADI','AAIC','AAIC^B','AAIC^C','AAIN']
```

```python
df = pd.DataFrame()

for i in symbol1:
    ticker = yf.Ticker(i)

    # Set the start and end dates for data retrieval
    start_date = datetime(2012, 1, 1)
    end_date = datetime(2016, 12, 31)

    historical_data = ticker.history(start=start_date, end=end_date, interval='1d')

    if historical_data.empty:
        continue

    historical_data['Symbol']=i
    df = df.append(historical_data)
    
df = df.reset_index()
df

df.to_csv('/Users/hp/SAD_Finance/stock_1.csv',index=False)
```

```python
#MYTE - "'float' object has no attribute 'upper'" 
unique_count1 = df['Symbol'].nunique()
print(unique_count1)
```
